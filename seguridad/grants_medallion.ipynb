## Creando los grupos

```sql
CREATE WIDGET TEXT catalog_name DEFAULT 'smartdata_project';
CREATE WIDGET TEXT schema_bronze DEFAULT 'bronze';
CREATE WIDGET TEXT schema_silver DEFAULT 'silver';
CREATE WIDGET TEXT schema_gold DEFAULT 'gold';
CREATE WIDGET TEXT external_location DEFAULT 'extl-raw';
CREATE WIDGET TEXT catalog_federated DEFAULT 'federated_sql_catalog';
```

### Creamos grupos para Data Analysts, Scientists y Engineers

## Agregando miembros a los grupos

## Data Engineers

```sql
GRANT USE CATALOG ON CATALOG ${catalog_name} TO `DEs`;
GRANT SELECT ON CATALOG ${catalog_name} TO `DEs`;
GRANT USE SCHEMA, CREATE, SELECT ON SCHEMA ${catalog_name}.${schema_bronze} TO `DEs`;
GRANT USE SCHEMA, CREATE, SELECT ON SCHEMA ${catalog_name}.${schema_silver} TO `DEs`;
GRANT USE SCHEMA, CREATE, SELECT ON SCHEMA ${catalog_name}.${schema_gold} TO `DEs`; 
GRANT READ FILES, WRITE FILES ON EXTERNAL LOCATION `${external_location}` TO `DEs`
```

## Data Analyst

```sql
GRANT USE CATALOG ON CATALOG ${catalog_name} TO `DAs`;
GRANT USE SCHEMA, CREATE TABLE, SELECT ON SCHEMA ${catalog_name}.${schema_gold} TO `DAs`;
```

## Data Scientist

```sql
GRANT USE CATALOG ON CATALOG ${catalog_name} TO `DSs`;
GRANT USE SCHEMA, CREATE TABLE, SELECT ON SCHEMA ${catalog_name}.${schema_silver} TO `DSs`;
GRANT USE SCHEMA, CREATE TABLE, SELECT ON SCHEMA ${catalog_name}.${schema_gold} TO `DSs`;
```
