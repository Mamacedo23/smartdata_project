## Creando los grupos

### Definimos una función para crear grupos cuando no existen

```sql
%python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()

def create_group_if_not_exists(group_name):
    existing = list(w.groups.list(filter=f"displayName eq '{group_name}'"))
    if not existing:
        w.groups.create(display_name=group_name)
        print(f"Grupo '{group_name}' creado")
    else:
        print(f"Grupo '{group_name}' ya existe, omitiendo")

def create_user(email):
    # Crear usuario en el workspace
    try:
        user = w.users.create(user_name=email)
        print(f"Usuario '{email}' creado")
    except Exception:
        user = next(w.users.list(filter=f"userName eq '{email}'"))
        print(f"Usuario '{email}' ya existe")
```

### Creamos grupos para Data Analysts, Scientists y Engineers

```sql
%python
grupos = ['de','da','ds']
for group_name in grupos:
    create_group_if_not_exists(group_name)
```

## Agregando miembros a los grupos

```sql
%python
users = ['mikoto@accelidol2gmail.onmicrosoft.com','rafa@accelidol2gmail.onmicrosoft.com','misaka@accelidol2gmail.onmicrosoft.com']
for user in users:
    create_user(user)
```
