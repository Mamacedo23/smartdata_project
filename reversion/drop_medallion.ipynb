## Eliminando los esquemas

```sql
CREATE WIDGET TEXT catalog_name DEFAULT 'smartdata_project';
CREATE WIDGET TEXT raw_name DEFAULT 'raw';
CREATE WIDGET TEXT bronze_name DEFAULT 'bronze';
CREATE WIDGET TEXT silver_name DEFAULT 'silver';
CREATE WIDGET TEXT gold_name DEFAULT 'gold';
```

```sql
DROP SHARE IF EXISTS delta_smartdata_project;
DROP RECIPIENT IF EXISTS users_smartdata_project;
```

```sql
DROP SCHEMA IF EXISTS ${catalog_name}.${raw_name} CASCADE;
DROP SCHEMA IF EXISTS ${catalog_name}.${bronze_name} CASCADE;
DROP SCHEMA IF EXISTS ${catalog_name}.${silver_name} CASCADE;
DROP SCHEMA IF EXISTS ${catalog_name}.${gold_name} CASCADE;
```

```sql
DROP CATALOG IF EXISTS ${catalog_name} CASCADE;
```

```sql
DROP EXTERNAL LOCATION IF EXISTS `extl-raw`;
```
