```python
dbutils.widgets.text('sales_date','1900-01-01')
dbutils.widgets.text('persons_date','1900-01-01')
dbutils.widgets.text('products_date','1900-01-01')
dbutils.widgets.text('catalog_name','smartdata_project')
dbutils.widgets.text('schema_name','raw')
```

```python
sales_date = dbutils.widgets.get('sales_date')
persons_date = dbutils.widgets.get('persons_date')
products_date = dbutils.widgets.get('products_date')
catalog_name = dbutils.widgets.get('catalog_name')
schema_name = dbutils.widgets.get('schema_name')
```

```python
import json

WATERMARK_PATH = f"/Volumes/{catalog_name}/{schema_name}/autoloader/watermarks.json"

# Leer valores de cada task upstream
updates = {k: v for k, v in {
    "sales":    sales_date,
    "persons":  persons_date,
    "products": products_date,
}.items() if v is not None}
```

```python
# Actualizar JSON de una sola vez
with open(WATERMARK_PATH, "r") as f:
    watermarks = json.load(f)

for entry in watermarks:
    if entry["table"] in updates:
        entry["last_updated"] = updates[entry["table"]]

with open(WATERMARK_PATH, "w") as f:
    json.dump(watermarks, f, indent=2)

print("✅ Watermarks actualizados:", updates)
```
